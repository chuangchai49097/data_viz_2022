HW (ให้ใช้เฉพาะที่อาจารย์สอนไปแล้วในวิชานี้)

1.   คำนวณ อายุเฉลี่ย ของผู้ชาย ของข้อมูลทั้งหมด
2.   คำนวณ อายุเฉลี่ย ของผู้ชาย ของผู้ป่วยในจังหวัดขอนแก่น

3. หาจำนวนผู้ป่วยที่เป็นคน "**คนต่างชาติเดินทางมาจากต่างประเทศ**"

````
this_data = data_covid[['sex','age','province_of_onset']] #####ดึงข้อมูลตารางมา
````

````
this_data ######### out put
````

````
m = this_data[this_data['sex']=='ชาย']        ########### จากตารางใน colum sex เอาแต่เพศชาย
m['age']                                     ########## out put
````

### ข้อ 1

````
sum = 0            ####### sum = 0 เพื่อเตรียมค่าก่อนบวก เป็นตัวแปรที่รอรับค่า อายุ
N = 0                 ####### N = 0 เพื่อเตรียมค่าก่อนบวก เป็นตัวแปรที่รอรับค่า จำนวนตัวแปร
for a in m['age']:      ####### เอาค่าข้อมูลอายุเข้ามาคำนวน
    
    if a > 0:           ############### ให้ค่า > 0 เพื่อใหค่าเป็นค่าปกติไม่มีค่าติดลบ
        sum += a          ##############  รวมอายุจากตาราง
        N += 1              #############  รวมจำนวนของข้อมูล จากcolumn

print(f'อายุเฉลี่ย ของ ผู้ป่วยชายเท่ากับ{sum/N}คน') ####### print เป็น out put
````

### ข้อ 2

````
mkk= m[m['province_of_onset']=='ขอนแก่น']            ################ เหมือนข้อ 1 แต่เปลี่ยน ในตารางเอาแค่ จ.ขอนแก่น

sum = 0                                             
N = 0
for a in mkk['age']:                                
    
    if a > 0:                                       
        sum += a                                    
        N += 1                                      

print(f'อายุเฉลี่ย ของ ผู้ป่วยชายในจังหวัดขอนแก่นเท่ากับ {sum/N} คน') ###### out put
````
