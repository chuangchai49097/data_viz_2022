# Import Data

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter([1,-1],[2,-1])
```

```python
plt.scatter(df['PetalLength'],df['SepalLength'])
```

```python
plt.scatter(df['PetalLength'],df['SepalLength'],c='r')
```

## ระบุสีให้แต่ละจุด

```python
set(df['Name'])
```

### แทนค่าชื่อดอกไม้ด้วยตัวอักษรแทนสี (r g b c m y k w)

```python
df2 = df.replace({'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'})
df2.head()
```

```python
plt.scatter(df['PetalLength'],df['SepalLength'],c=df2['Name'])
```

## เพิ่มรายละเอียดกราฟ

### ชื่อแกน

```python
plt.scatter(df2['PetalLength'],df2['SepalLength'],c=df2['Name']) # วาดกราฟ
plt.xlabel('PetalLength')   # ใส่ชื่อแกน x
plt.ylabel('SepalLength')   # ใส่ชื่อแกน y
```

### ชื่อ marker

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50], label='setosa') # วาดกราฟ
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100], label='versicolor') # วาดกราฟ
plt.scatter(df2['PetalLength'][100:],df2['SepalLength'][100:],c=df2['Name'][100:], label='virginica') # วาดกราฟ
plt.legend()  # แสดงชื่อที่ plot
plt.xlabel('PetalLength')   # ใส่ชื่อแกน x
plt.ylabel('SepalLength')   # ใส่ชื่อแกน y
```

### ชื่อกราฟ

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50], label='setosa') # วาดกราฟ
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100], label='versicolor') # วาดกราฟ
plt.scatter(df2['PetalLength'][100:],df2['SepalLength'][100:],c=df2['Name'][100:], label='virginica') # วาดกราฟ
plt.legend()
plt.xlabel('PetalLength')   # ใส่ชื่อแกน x
plt.ylabel('SepalLength')   # ใส่ชื่อแกน y
plt.title('comparing petal&sepal length of different iris flowers');
```

### ใช้ ขนาดของ marker ในการแสดงค่าใน column

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50], label='setosa', s=20*df2['PetalWidth'][:50]) # วาดกราฟ
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100], label='versicolor', s=20*df2['PetalWidth'][50:100]) # วาดกราฟ
plt.scatter(df2['PetalLength'][100:],df2['SepalLength'][100:],c=df2['Name'][100:], label='virginica', s=20*df2['PetalWidth'][100:]) # วาดกราฟ
plt.legend()
plt.xlabel('PetalLength')   # ใส่ชื่อแกน x
plt.ylabel('SepalLength')   # ใส่ชื่อแกน y
plt.title('comparing petal&sepal length of different iris flowers');
```

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50], label='setosa', s=40*df2['PetalWidth'][:50]) # วาดกราฟ
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100], label='versicolor', s=40*df2['PetalWidth'][50:100]) # วาดกราฟ
plt.scatter(df2['PetalLength'][100:],df2['SepalLength'][100:],c=df2['Name'][100:], label='virginica', s=40*df2['PetalWidth'][100:]) # วาดกราฟ
plt.legend()
plt.xlabel('PetalLength')   # ใส่ชื่อแกน x
plt.ylabel('SepalLength')   # ใส่ชื่อแกน y
plt.title('comparing petal&sepal length of different iris flowers');
```

### ปรับความโปร่งใสของ marker เพื่อให้มองเห็นจุดที่ถูกบัง (Bubble Chart)

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50], alpha=0.5, label='setosa', s=40*df2['PetalWidth'][:50]) # วาดกราฟ
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100], alpha=0.5, label='versicolor', s=40*df2['PetalWidth'][50:100]) # วาดกราฟ
plt.scatter(df2['PetalLength'][100:],df2['SepalLength'][100:],c=df2['Name'][100:], alpha=0.5, label='virginica', s=40*df2['PetalWidth'][100:]) # วาดกราฟ
plt.legend()
plt.xlabel('PetalLength')   # ใส่ชื่อแกน x
plt.ylabel('SepalLength')   # ใส่ชื่อแกน y
plt.title('comparing petal&sepal length of different iris flowers');
```

### เปลี่ยน หน้าตา marker เพื่อความสวยงาม
