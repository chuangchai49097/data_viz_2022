# Parallel Coordinates

```python
pd.plotting.parallel_coordinates(df,'Name') # Sort ตารางค่าแต่ละ Corlumn เทียบค่ากัน โดยใช้ parallel coordinates # โดยกำหนดค่าเป็น (ชื่อตัวแปรข้อมูลที่ต้องการดู,ชื่อ column ที่ต้องการ)
```

### select columns to show

```python
pd.plotting.parallel_coordinates(df,'Name',cols = ['SepalLength','PetalLength']);
```

### change color

https://matplotlib.org/stable/tutorials/colors/colors.html

```python
pd.plotting.parallel_coordinates(df,'Name',color = ['r','y','k']) # ใช้สีจากค่าสีที่มีอยู่แล้วทั่วไป
```

```python
pd.plotting.parallel_coordinates(df,'Name',color = [[0.5,0.5,0.5],[0,0,0.8],[0.2,0.4,0.7]]) # ใช้การผสมสีเอง จากค่าสีที่ต้องการ
```

```python
pd.plotting.parallel_coordinates(df,'Name',color = ['#FF00E0','#00FFE2','#ECFF00']) # การใช้ค่าสี โดยการใช้ code สี
```

### Specify color by colormap

https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
pd.plotting.parallel_coordinates(df,'Name',colormap = 'Set1') # การเลือกใช้สีตาม colormap โดยที่ถ้ามีข้อมูล 3 ชนิดก็จะสุ่มสีโดยการสุ่ม หัว,กลาง,ท้าย ของ colormap
```

วาดกราฟ parallel coordinate เปรียบเทียบ [อายุ และ เพศ] โดยใช้ class เป็น risks (เลือกมาแค่ 3 risks)

```python
import pandas as pd # import ฟังก์ชัน pandas
```

```python
from google.colab import drive # เชื่อม drive ของเรา ถ้าเชื่อมสำเร็จจะขึ้นคำว่าMounted at /content/drive
drive.mount('/content/drive')
import os # จัดการเกี่ยวกับไฟล์ต่างๆ ในโฟล์เดอร์
path = '/content/drive/My Drive/dataviz_2021_data' # ชี้ว่า data เราอยู่โฟล์เดอร์ไหน
```

```python
data_file_path = os.path.join(path,'pm-20-jan-2021.csv') # เอาที่อยู่ของไฟล์เชื่อมกับที่อยู่ของไฟล์เรา โดยกำหนดชื่่อไฟล์ที่เราต้องการ
data_covid = pd.read_csv(data_file_path)
```

```python
data_covid
```

```python
data_covid.isnull().any() # เช็คดูต่า missing
```

```python
data_covid['age'].mean() # หาค่า mean ของ age
```

```python
avg_age = data_covid['age'].mean() 
data_covid['age'] = data_covid['age'].fillna(avg_age) # แทนค่าเฉลี่ยของอายุลงในค่าวาง
```

```python
data_covid.isnull().any() # เช็คค่า missing ของ age
```

```python
data_covid['sex'].unique() # เช็คดูว่า sex เก็บค่าอะไรบ้าง
```

```python
data_covid['sex'].mode() # หาค่าที่พบมากที่สุดของ sex โดยใช้ฐานนิยม
```

```python
data_covid = data_covid.fillna(value={'sex':'หญิง'}) # แทนค่า 'หญิง' แทนค่า missing
data_covid.head()
```

```python
data_covid.head()
```

```python
data_covid['sex'].replace({'ชาย':'0','หญิง':'99'}, inplace=True) # เปลี่ยนค่าโดยกำหนดให้ ชาย เป็น 0 และ หญิง เป็น 99
```

```python
data_covid.head()
```

```python
data_covid.isnull().any() # เช็คค่า missing ของ sex
```

```python
data_covid['risk'] = data_covid['risk'].fillna('ไม่ได้ระบุข้อมูล') # แทนค่า nan ใน column 'risk' ด้วย 'ไม่ได้ระบุข้อมูล'
```

```python
data_covid.head()
```

```python
data_covid.isnull().any() # เช็คค่า missing ของ risk
```

```python
data = data_covid[['risk','age','sex']] # เลือกมาเฉพาะข้อมูลที่ต้องใช้ คือ risk,age,sex
data
```

```python
data['risk'].unique() # ดูว่า risk เก็บค่าอะไรบ้าง
```

```python
data_risk = data[data['risk']== 'สนามมวย' ] # เลือกค่าที่จะใช้
data_risk
```

```python
data_risk1 = data[data['risk']== 'สถานบันเทิง' ] # เลือกค่าที่จะใช้
data_risk1
```

```python
data_risk2 = data[data['risk']== 'สัมผัสผู้ป่วยยืนยัน' ] # เลือกค่าที่จะใช้
data_risk2
```

```python
data_cluster = pd.concat([data_risk,data_risk1,data_risk2]) # นำค่าที่จะใช้ทั้ง 3 ค่ามารวมกันเป็นตารางเดียว
data_cluster
```

```python
data_cluster['risk'].replace({'สนามมวย':'boxing stadium','สถานบันเทิง':'entertainment place','สัมผัสผู้ป่วยยืนยัน':'touch the patient'}, inplace=True) # เปลี่ยนชื่อใน column risk จากภาษาไทย เป็นภาษาอังกฤษ
```

```python
data_cluster
```

```python
pd.plotting.parallel_coordinates(data_cluster,'risk',color = ['#FF00E0','#00FFE2','#ECFF00'])  #plot ข้อมูลเลือกสีในการ plot
```
