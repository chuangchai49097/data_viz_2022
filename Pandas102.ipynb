# ต่อตาราง

* ต่อในแนวแกน Y (เพิ่มจำนวนข้อมูล)
* ต่อในแนวแกน X (เพิ่มรายละเอียดของข้อมูล)

## ต่อในแนวแกน Y

```python
BKK = data_df[data_df['province_of_onset'] == 'กรุงเทพมหานคร']
BKK.head()
```

```python
KKC = data_df[data_df['province_of_onset'] == 'ขอนแก่น']
KKC.head()
```

### ต่อคารางใฟ้เป็นข้อมูล 2 จังหวัด

```python
dataBKKKKC = pd.concat([BKK,KKC])       # .concat เอาไว้ต่อตาราง(ลงด้านล่าง) 
```

```python
dataBKKKKC
```

### ถ้าเอาตารางที่ไม่เหมือนกันมา concat กัน

```python
BKK2 = data_df[data_df['province_of_onset'] == 'กรุงเทพมหานคร'][['sex','age','nationality']]
BKK2.head()
```

```python
KKC2 = data_df[data_df['province_of_onset'] == 'ขอนแก่น'][['age','nationality','sex']]
KKC2.head()
```

```python
pd.concat([BKK2,KKC2])      # ชื่อคอลัมน์สลับที่กันเอามาต่อกันได้
```

```python
KKC3 = data_df[data_df['province_of_onset'] == 'ขอนแก่น'][['age','nationality','sex','risk']]
KKC3.head()
```

```python
pd.concat([BKK2,KKC3])
```

## ต่อในแนวแกน X

```python
personal_data = data_df[['No.','sex','age','nationality']]
personal_data.head()
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']].sample(frac=1)
covid_data.head()
```

### เอาตารางสองอันมาต่อกัน

```python
personalcovid_data = personal_data.merge(covid_data)
personalcovid_data.head()
```

```python
personalcovid_data = covid_data.merge(personal_data)
personalcovid_data.head()
```

# Save Table

```python
personalcovid_data.to_csv(os.path.join(path,'my_processed_data.csv') )      # save ลง drive
```
